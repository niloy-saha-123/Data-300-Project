```python
import sys
import os
import pandas as pd
import numpy as np

# Add the parent directory to the Python path
sys.path.append(os.path.abspath('..'))

# Now Python can find and import your shared plotting style
import src.utils.plotting
```

```python
# Load the processed datasets
portfolio_df = pd.read_parquet('../data/processed/portfolio_clean.parquet')
profile_df = pd.read_parquet('../data/processed/profile_clean.parquet')
transcript_df = pd.read_parquet('../data/processed/transcript_flat.parquet')

print("Data loaded successfully!")
```

```python
print("--- Data Shapes ---")
print(f"Portfolio shape: {portfolio_df.shape}")
print(f"Profile shape: {profile_df.shape}")
print(f"Transcript shape: {transcript_df.shape}")
```

```python
datasets = {
    "Portfolio": portfolio_df,
    "Profile": profile_df,
    "Transcript": transcript_df
}

for name, df in datasets.items():
    print(f"\n--- {name} Dataset ---")
    
    # Create a summary dataframe for types and nulls
    summary = pd.DataFrame({
        'Data Type': df.dtypes,
        'Missing Values': df.isnull().sum(),
        '% Missing': (df.isnull().sum() / len(df) * 100).round(2)
    })
    
    display(summary)
```

```python
print("\n--- Unique Identifiers ---")
# The transcript ties customers and offers together
unique_customers = transcript_df['person'].nunique()
# For offers, we can check both the portfolio and the events
unique_offers_portfolio = portfolio_df['offer_id'].nunique()

print(f"Total Unique Customers (in transcript): {unique_customers:,}")
print(f"Total Unique Offers (in portfolio): {unique_offers_portfolio}")
```

```python
print("\n--- Transcript Event Type Frequency ---")
event_counts = transcript_df['event'].value_counts()
event_percentages = transcript_df['event'].value_counts(normalize=True) * 100

event_summary = pd.DataFrame({
    'Count': event_counts,
    'Percentage (%)': event_percentages.round(2)
})

display(event_summary)
```
